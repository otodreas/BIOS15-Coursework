### **Exercise 1.1**
Use non-parametric bootstrapping to derive a 95% confidence interval for the CV of a variable. Start by writing a function that computes the CV for a variable(see the Appendix for a brief introduction to writing functions in R). Then, simulate a random variable and write a loop that samples many times from this variable and computes the CV.

```r
compute_cv <- function(u_true, sd_true) {
  cv_true <- sd_true/u_true
  return(round(cv_true*100, 2))
}

simulate_cv <- function(u_true, sd_true, n) {
  n <- 1000
  data <- rnorm(n, u, sd)
  sd_sim <- sd(data)
  u_sim <- mean(data)
  cv_sim <- sd_sim/u_sim
  return(round(cv_sim*100, 2))
}

u_true <- 23  # true mean of dataset
sd_true <- 10  # true standard deviation of dataset
n <- 100  # size of simulated dataset
cat("Theoretical CV (%):", compute_cv(u_true, sd_true), "\n", sep=" ")
cat("Simulated CVs (%):", simulate_cv(u_true, sd_true, n), sep=" ")
```
